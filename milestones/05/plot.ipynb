```python
import matplotlib.pyplot as plt

# File path
file_path = 'energy.txt'

# Initialize lists to hold time and energy values
time = []
energy = []

# Read the data from the file
with open(file_path, 'r') as file:
    for line in file:
        t, e = line.split()
        time.append(float(t))
        energy.append(float(e))

# Plot the data
plt.figure(figsize=(15, 10))
plt.plot(time, energy, label='Total Energy')

# Labels and title
plt.xlabel('Time ($t$)')
plt.ylabel('Total Energy ($E_{{pot}}$)')
plt.title('Total Energy as a Function of Time')
plt.legend()
plt.grid(True)
plt.show()
```
