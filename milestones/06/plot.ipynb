```python
import matplotlib.pyplot as plt
import pandas as pd

simulation_times_cutoff= [1.9986, 6.2961, 29.9082, 68.7376, 131.315, 214.809    ]
simulation_times= [1.11875, 4.38685, 29.1829, 116.306, 249.649, 428.411    ]
atom_counts = [ 100, 200, 500, 1000, 1500, 2000]

plt.figure(figsize=(12, 8))
plt.plot(atom_counts, simulation_times_cutoff, marker='o')
plt.text(0.05, 0.95, '(b)', transform=plt.gca().transAxes, fontsize=25, va='top', ha='left')
plt.xlabel('Number of Atoms', fontsize=20)
plt.ylabel('Simulation Time (seconds)', fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.title('Simulation Time as a Function of System Size with cutoff', fontsize=25)
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 8))
plt.plot(atom_counts, simulation_times, marker='o')
plt.text(0.05, 0.95, '(a)', transform=plt.gca().transAxes, fontsize=25,  va='top', ha='left')
plt.xlabel('Number of Atoms', fontsize=20)
plt.ylabel('Simulation Time (seconds)', fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.title('Simulation Time as a Function of System Size witout cutoff', fontsize=25)
plt.grid(True)
plt.show()
```
