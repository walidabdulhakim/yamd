```python
import matplotlib.pyplot as plt

# File path
file_path = 'energy04.txt'

# Initialize lists to hold time and energy values
time = []
energy = []

# Read the data from the file
with open(file_path, 'r') as file:
    for line in file:
        t, e = line.split()
        time.append(float(t))
        energy.append(float(e))

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(time, energy, label='Total Energy')

# Labels and title
plt.xlabel('Time steps ($t$)', fontsize=14)
plt.ylabel('Total Energy ($E_{{pot}}$)', fontsize=14)
plt.title('Total Energy as a Function of Time')
plt.legend()
plt.grid(True)
plt.show()
```
