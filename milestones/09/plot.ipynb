```python
import pandas as pd
import matplotlib.pyplot as plt
from sympy import symbols, latex
from sympy import Eq
# Define symbols
epsilon= symbols('epsilon')

# Create the equation
equation = Eq(epsilon,1)

# Load the CSV data
file_path = 'whisker_large_reading.csv'  # Replace with the actual file path
data = pd.read_csv(file_path)

# Calculate stress (force_zz / cross-sectional area)
data['stress_zz'] = data['force_zz'] / (data['volume_x'] * data['volume_y'])

# Calculate strain (change in z-direction volume relative to the initial volume)
# Use .iloc[0] to ensure we get the first value by position, not by index
initial_volume_z = data['volume_z'].iloc[0]
data['strain_zz'] = (data['volume_z'] - initial_volume_z) / initial_volume_z

# Plot the stress-strain curve
plt.figure(figsize=(8, 6))
plt.plot(data['strain_zz'], data['stress_zz'], label=f"${latex(equation)}*10^8$, T = 0 K", color='orange')
plt.xlabel('Strain ()', fontsize=14)
plt.ylabel('Stress () [Pa]', fontsize=14)
plt.title('Stress-Strain Curve')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend()
plt.grid(True)
plt.show()
```
